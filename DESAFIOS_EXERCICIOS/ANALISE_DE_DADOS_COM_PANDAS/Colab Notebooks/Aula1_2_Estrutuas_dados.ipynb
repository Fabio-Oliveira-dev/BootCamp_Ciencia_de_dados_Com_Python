# **Tuplas**


```python
# As tuplas usam parênteses como sintaxe
tp = ('Banana', 'Maçã', 10, 50)
```

```python
tp[0]
```

```python
# Difirentes das listas, as tuplas são imutáveis, quer dizer que não podemos alterar os seus elementos
tp[0] = 'Laranja'
```

```python
tp.count('Maçã')
```

```python
tp[0:2]
```

# **Dicionários**

```python
# Para adiconar um dicionário utilizamos as chaves {}
dc = {'Maçã':20, 'Banana':10, 'Laranja':15, 'Uva':5} # Dicionários trabalham com conceitos de chaves e valor
```

```python
dc
```

```python
# Acessando um valor do dicionário através da chave
dc['Maçã']
```

```python
# Atualizando o valor da maçã
dc['Maçã'] = 25
dc
```

```python
# retornando todas as chaves do dicionário
dc.keys()
```

```python
# Retornando valores do dicionário
dc.values()
```

```python
# Verificando se já existe uma chave no dicionário e caso não exista inserir
dc.setdefault('Limão', 22)
```

```python
dc
```
