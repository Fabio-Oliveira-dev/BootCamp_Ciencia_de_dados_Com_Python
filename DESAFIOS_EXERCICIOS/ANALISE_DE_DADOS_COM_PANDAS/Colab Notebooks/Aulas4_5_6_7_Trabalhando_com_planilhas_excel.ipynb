# **# Trabalhando com planilhas do excel**

```python
# Importando a biblioteca pandas
import pandas as pd
```

```python
# Leitura dos arquivos
df1 = pd.read_excel('Aracaju.xlsx')
df2 = pd.read_excel('Fortaleza.xlsx')
df3 = pd.read_excel('Natal.xlsx')
df4 = pd.read_excel('Recife.xlsx')
df5 = pd.read_excel('Salvador.xlsx')
```

```python
df1.head()
```

```python
# Juntando todos os arquivos
df = pd.concat([df1, df2, df3, df4, df5])
```

```python
# Exibindo as 5 primneiras linhas
df.head()
```

```python
# Exibindo as 5 ultimas linhas
df.tail()
```

```python
# Pegar uma amostra de 5 linhas de cada arquivo de dados
df.sample(5)
```

```python
# Verificando o tipo de dados de cada coluna
df.dtypes
```

```python
# Alterando o tipo de dado da coluna LojaID
df['LojaID'] = df['LojaID'].astype('object')
```

```python
df.dtypes
```

```python
df.head()
```

# **Tratando valores faltantes**

```python
# Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
# Substituindo os valores nulos pela média
#df.['Vendas'].fillna(df['Vendas'].mean(), inplace=True)
```

```python
df['Vendas'].mean()
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
# Substituindo os valores nulos por 0
df['Vendas'].fillna(0, inplace=True)
```

```python
# Apagando linhas com valores nulos
df.dropna(inplace=True)
```

```python
# Apagando linhas com valores nulos com base em apenas 1 coluna
df.dropna(subset['vendas'], inplace=True)
```

```python
# Removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(row='all', inplace=True)
```

# **Criando colunas novas**

```python
# Criando a coluna de receita
df['Receita'] = df['Vendas'].mul(df['Qtde'])
```

```python
df.head()
```

```python
# Caso eu não tenha a coluna quantidade eu posso fazer o calculo dessa forma
df['Receita/Vendas'] = df['Receita'] / df['Vendas']
```

```python
df.head()
```

```python
# Retornando a maior receta
df['Receita'].max()
```

```python
# Retornando a menor receita
df['Receita'].min()
```

```python
# Detectando qual loja teve mais receita. Top 3 de que vendeu em receita
df.nlargest(3, 'Receita')
```

```python
# Retornando as top 3 piores receita
df.nsmallest(3, 'Receita')
```

```python
# Argumento por cidade
df.groupby('Cidade')['Receita'].sum()
```

```python
# Ordenando os conjuntos de dados
df.sort_values('Receita', ascending=False).head(10)
```

# **Trabalhando com datas**

```python
 # Transformando a coluna de data em tipo inteiro
 df['Data'] = df['Data'].astype('int64')
```

```python
# Verificando o tipo de dado de cada coluna
df.dtypes
```

```python
# Transformando a coluna de data em data
df['Data'] = pd.to_datetime(df['Data'])
```

```python
df.dtypes
```

```python
# Agrupamento por ano
df.groupby(df['Data'].dt.year)['Receita'].sum()
```

```python
# Criando uma nova coluna com o ano
df['Ano_venda'] = df['Data'].dt.year
```

```python
df.sample(5)
```

```python
# Extraindo o mês e o dia
df['Mês_venda'], df['Dia_venda'] = (df['Data'].dt.month, df['Data'].dt.day)
```

```python
df.sample(5)
```

```python
# Retornando a data mais antiga
df['Data'].min()
```

```python
# Calculando a diferença de dias
df['Diferença_dias'] = df['Data'] - df['Data'].min()
```

```python
df.sample(5)
```

```python
# Criando a coluna de trimestre
df['Trimestre_venda'] = df['Data'].dt.quarter
```

```python
df.sample(6)
```

```python
# Filtrando as vendas de 2019 do mês de março
Vendas_março_2019 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
```

```python
Vendas_março_2019.sample(20)
```
